## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
#Sub Problem 1: How to aggregate the amount_spent for unique customers?

amount_spent = orders[['CustomerID','amount_spent','Country']].groupby(['CustomerID','Country']).sum()

amount_spent

```

```python
#Sub Problem 2: How to select customers whose aggregated amount_spent is in a given quantile range?
#Sub Problem 3: How to label selected customers as "VIP" or "Preferred"

quantile75 = float(amount_spent['amount_spent'].quantile([0.75]))
quantile95 = float(amount_spent['amount_spent'].quantile([0.95]))

vip_customers = amount_spent[amount_spent.loc[:,'amount_spent'] >= quantile95]


preferred_customers = amount_spent[(amount_spent.loc[:,'amount_spent'] >= quantile75) &
                                   (amount_spent.loc[:, 'amount_spent']<quantile95)]

top_customers = pd.concat([vip_customers,preferred_customers])

top_customers
```

## Q2: How to identify which country has the most VIP Customers?

```python
top_customers["Customer_Status"] = ["VIP" if amount >= quantile95 else "Preferred" for amount in top_customers["amount_spent"]]

vip_customers.columns

# vip_customers = vip_customers.reset_index()

vip_customers.columns

vip_customers_ordered = vip_customers.groupby(['Country'])['CustomerID'].count().sort_values(ascending=False)

df_vip_customers = vip_customers_ordered.to_frame().reset_index()

df_vip_customers.max()
```
